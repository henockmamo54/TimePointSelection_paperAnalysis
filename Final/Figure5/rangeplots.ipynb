```python
from matplotlib import pyplot as plt
import matplotlib.patches as patches
import pandas as pd
import math
```

```python
ph=1.5574E-4
def computeI0_t(I0_0,pw,neh,I0_asymp,t,k):
    
    return I0_asymp + (I0_0-I0_asymp)*math.exp(-k*t)
    
```

```python
seq="AQTAHIVLEDGTK" #CPSM_MOUSE
I0_0=45.62/100
neh=25
pw=0.046
I0_asymp= I0_0 * ( (1-pw/(1-ph))**neh )
```

```python

times=range(0,16)
slow=[computeI0_t(I0_0,pw,neh,I0_asymp,t,k=0.06) for t in times]
fast=[computeI0_t(I0_0,pw,neh,I0_asymp,t,k=0.12) for t in times]
```

```python
# =======================> slow

fig, ax = plt.subplots(figsize=(8,5))

uperlimt=I0_0 - 0.031
lowerlimt=0.65*I0_0 +0.35*I0_asymp
 

# Create a rectangle patch
rect=patches.Rectangle(( 2 - 0.4, lowerlimt-0.001),
                        5  +0.7, 
                        uperlimt-lowerlimt +0.001, facecolor="g", alpha=0.2)


ax.plot(times,[lowerlimt]*len(times),c='m',ls=':')
ax.plot(times,[uperlimt]*len(times),c='r',ls=':') 

ax.scatter(times,slow,c='k',alpha=0.7 )

ax.add_patch(rect)


ax.text(12, uperlimt, "Lower Limit (theo.) = " +r"$0.65 * I_0(0) + 0.35 * I_0^{asmyp}$", ha='center', va='bottom', fontsize=8, color='r')
ax.text(14, lowerlimt, r"Upper Limit (theo.) = $I_0(0) - 0.031$", ha='center', va='bottom', fontsize=8, color='m')

plt.gca().spines['top'].set_visible(False)
plt.gca().spines['right'].set_visible(False)
plt.ylim([I0_asymp-0.01,I0_0+0.02])

plt.xlabel("Labeling Duration (Days)")
ax.set_ylabel(r"$I_0(t) = I_0^{asmyp} + (I_0(0) - I_0^{asmyp})e^{-kt}$")
# plt.legend(frameon=False, loc='upper right')


plt.savefig("_Figure5_006.jpeg",dpi=900)

plt.show()
```
